# **Data Preprocessing**

## **Handling missing data**
If any compounds has missing value for the **standard_value** column then drop it

```
df2 = df[df.standard_value.notna()]
df2 = df[df.value.notna()]
df2
```

```
fix = []

for i in df2.standard_value:
    fix.append(float(i))
```

```
# define active, inactive, intermediate compounds

STATUS = []

for i in fix:
    if i <=1000:
        STATUS.append("active") #active
        
    elif i >=10000:
        STATUS.append("inactive") #inactive
        
    else:
        STATUS.append("intermediate") #intermediate
```

```
df2 = df2.drop('value', 1)
df2['value'] = fix
df2['STATUS'] = STATUS
df2.head(3)
```

```
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value', 'STATUS']
df3 = df2[selection]
df3
```

```
selection1 = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df4 = df2[selection1]
df4
```

```
# find the duplicate and unique molecule_chembl_id
import pandas as pd

df2_dup = pd.concat(g for _, 
                    g in df2.groupby("molecule_chembl_id") 
                      if len(g) > 1)
df2_non = df2.loc[~df3.index.isin(df2_dup.index)]

print (str(len(df2_non)), 
       str(len(df2_dup)), 
       str(len(df2_dup)+len(df2_non)))
```

```
len(df2_non), len(df2_non['molecule_chembl_id'].unique())
```

```
df2_dup
```

```
# mean and std of all duplicate
import numpy as np
import pandas as pd
from time import time

from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix
import seaborn as sns
%matplotlib inline

mean_std = df2_dup.groupby(['molecule_chembl_id'], as_index=False).agg({'value':['mean', 'std']})
mean_std.head()
```

```
len(mean_std)
```

```
# merging the dupilicate chemblID on mean_std dataframe
df2_dup = df2_dup.merge(mean_std, on='molecule_chembl_id', how='inner')
len(df2_dup)
```

```
#keep only SD less thab 2SD for duplicate chemblID

df2_dup = df2_dup[(df2_dup[('value', 'std')] < 2)]
len(df2_dup), len(df2_dup['molecule_chembl_id'].unique())
```

```
df2_dup['select'] = (df2_dup['value']- df2_dup[('value', 'mean')]).abs()
#value_median = keep.groupby('chemblId')['value'].transform('median')

df2_dup = df2_dup.groupby(["molecule_chembl_id"]).apply(lambda x: x.sort_values(["select"], ascending = True)).reset_index(drop=True)
df2_dup.head(10)
```

```
# since we assign the mean value as value to all the duplicates chembleID
# we can remove the duplicate chemblID and only keep the first
df2_dup = df2_dup.drop_duplicates(subset='molecule_chembl_id', keep='first')
df2_dup
```

```
len(df2_dup), len(df2_dup['molecule_chembl_id'].unique())
```

```
df2_dup = df2_dup.drop('select', 1)
df2_dup = df2_dup.drop(('value', 'mean'), 1)
df2_dup = df2_dup.drop(('value', 'std'), 1)
```

```
df2_final = pd.concat([df2_non, df2_dup])
len(df2_dup), len(df2_non), len(df2_final)
```
