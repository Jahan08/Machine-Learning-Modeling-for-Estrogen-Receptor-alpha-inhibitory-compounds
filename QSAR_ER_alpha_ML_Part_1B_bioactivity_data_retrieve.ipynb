## **Search for Target protein**

### **Target search for ER-alpha**

two-thirds of breast cancers are positive for estrogen receptor (ERs), 
which is a predisposing factor for the growth of breast cancer cells
ERa represents a lucrative therapeutic target for breast cancer that has attracted wide interest in the search for inhibitory agents. 

```
# Target search for Estrogen receptor alpha
target = new_client.target
target_query = target.search('estrogen receptor alpha')
targets = pd.DataFrame.from_dict(target_query)
targets
```

### **Select and retrieve bioactivity data for *Estrogen Receptor Alpha* (forth entry) CHEMBL206**

We will assign the forth entry (which corresponds to the target protein, *Estrogen Receptor (Alpha*) to the ***selected_target*** variable 

```
selected_target = targets.target_chembl_id[3]
selected_target
```

Here, we will retrieve only bioactivity data for *estrogen receptor alpha* (CHEMBL206) that are reported as IC$_{50}$ values in nM (nanomolar) unit.

```
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```
df = pd.DataFrame.from_dict(res)
```

```
df = df[(df['assay_type'] == 'B')]
len(df)
```

```
# considering nM unit for IC50
df = df[(df['units'] == 'nM')]
len(df)
```

# **Data Preprocessing**

## **Handling missing data**
If any compounds has missing value for the **standard_value** column then drop it

```
df2 = df[df.standard_value.notna()]
df2 = df[df.value.notna()]
df2
```

```
fix = []

for i in df2.standard_value:
    fix.append(float(i))
```

```
# define active, inactive, intermediate compounds

STATUS = []

for i in fix:
    if i <=1000:
        STATUS.append("active") #active
        
    elif i >=10000:
        STATUS.append("inactive") #inactive
        
    else:
        STATUS.append("intermediate") #intermediate
```

```
df2 = df2.drop('value', 1)
df2['value'] = fix
df2['STATUS'] = STATUS
df2.head(3)
```

```
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value', 'STATUS']
df3 = df2[selection]
df3
```

```
selection1 = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df4 = df2[selection1]
df4
```

```
df3.to_csv('bioactivity_preprocessed_data_ERA_1.csv', index=False)
```
